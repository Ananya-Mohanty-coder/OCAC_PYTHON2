```python
#1. Python Program to Add Two Numbers , take input from user and show the input and output uisng .format
def add(x, y):
    return x+y
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
result = add(num1,num2)
print("you entered: {0} and {1})".format(num1,num2))
print("The sum of {0}and {1}is: {2}".format(num1,num2,result))

```

```python
 #2.Python Program to Find the Square Root using operator.
def square_root(number):
    return number ** 0.5

num = float(input("Enter a number to find its square root: "))
result = square_root(num)
print("The square root of", num, "is:", result)
```

```python
#3. Python Program to Calculate the Area of a Triangle.
def calculate_area(base, height):
    return 0.5 * base * height

base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))
result = calculate_area(base, height)
print("The area of the triangle is:", result)

```

```python
#4.Python Program to Convert Kilometers to Miles.
def kilometers_to_miles(kilometers):
    return kilometers * 0.621371
kilometers = float(input("Enter the distance in kilimeters: "))
miles = kilometers_to_miles(kilometers)
print(kilometers, "kilometers is equal to{miles} miles")

```

```python
#5.Python Program to Convert Celsius To Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius = float(input("Enter the temperature in Celsius: "))
fahrenheit = celsius_to_fahrenheit(celsius)
print(celsius, "Celsius is equal to", fahrenheit, "Fahrenheit")



```

```python
#6. Python Program to Find ASCII Value of Character.
def ascii_value(character):
    return ord(character)

character = input("Enter a character: ")
result = ascii_value(character)
print("The ASCII value of", character, "is:", result)



```
