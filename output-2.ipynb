```python
a = int(input("Enter the value of a: "))
b=int(input("Enter the value of b: "))
add=a+b
sub=a-b
mult=a*b
div=a/b
mod=a%b
print("ADD:", add)
print("Sum: {0}, Diff: {1}, Mult:{2},Mod: {4}".format(add,sub,mult,div,mod))
```

```python
print("Sum: {}, Diff: {}, Mult:{}, Div:{}, Mod: {}".format(add, sub, mult, div, mod))
```
